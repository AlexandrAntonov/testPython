```python
# В процентах
evbyusper = evbyus
for g in evbyus.columns:
    evbyusper['per'+g] = 100*evbyus[g]/int(totus[g])
evbyusper
```

```python
# Функция проверки p-value
def comp(event,group1,group2):
    alpha = 0.05
    gr = np.array([evbyus.loc[event,group1], evbyus.loc[event,group2]])
    tot = np.array([int(totus[group1]), int(totus[group2])])
    p1 = gr[0]/tot[0]
    p2 = gr[1]/tot[1]
    p_combined = (gr[0]+gr[1])/(tot[0]+tot[1])
    difference = p1-p2
    z_value = difference / mth.sqrt(p_combined * (1-p_combined) * (1/tot[0] + 1/tot[1]))
    distr = st.norm(0,1)
    p_value = (1 - distr.cdf(abs(z_value)))*2
    print('\n*** Проверяем событие {}, между группами {} и {} ***'.format(event,group1,group2))
    print('p-значение: ', p_value)
    if p_value<alpha:
        print('Отвергаем нулевую гипотезу: В группах есть различие')
    else:
        print('Не получилось отвергнуть нулевую гипотезу, нет оснований считать группы разными')
```
